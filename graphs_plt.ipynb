```python
import pandas as pd
import matplotlib.pyplot as plt
iris=pd.read_csv('/content/iris.csv')
x=iris['SepalWidthCm']
y=iris['SepalLengthCm']
plt.scatter(x,y)
plt.xlabel('SepalWidthCm')
plt.ylabel('SepalLengthCm')
plt.title('scatter plot')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
iris=pd.read_csv('/content/iris.csv')
x=iris['SepalWidthCm']
y=iris['SepalLengthCm']
color={'Iris-setosa':'red','Iris-virginica':'blue','Iris-versicolor':'green'}
species_color=iris['Species'].map(color)
plt.scatter(x,y,c=species_color)
plt.xlabel("SepalWidthCm")
plt.ylabel("SepalLengthCm")
plt.title("scatter plot")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.pyplot as plt
iris=pd.read_csv('/content/iris.csv')
x=iris['SepalWidthCm']
y=iris['SepalLengthCm']
species_counts=iris['Species'].value_counts()
plt.bar(species_counts.index,species_counts.values,color="green")
plt.title("Number of flowers by species")
plt.xlabel("species")
plt.ylabel("count")
plt.grid(axis='y')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
iris=pd.read_csv('/content/iris.csv')
plt.hist(iris['PetalLengthCm'], bins=10, color='lightgreen', edgecolor='black')
plt.title("Distribution of Petal Length")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
iris=pd.read_csv('/content/iris.csv')
x=iris['PetalLengthCm']
species = iris['Species'].unique()
data = [iris[iris['Species'] == s]['PetalLengthCm'] for s in species]
plt.boxplot(data, labels=species)
plt.title("Petal Length by Species")
plt.xlabel("Species")
plt.ylabel("Petal Length (cm)")
plt.grid(True)
plt.show()
```
